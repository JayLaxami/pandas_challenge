```python
# Dependencies and Setup
import pandas as pd

# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools_csv = pd.read_csv(school_data_to_load)
students_csv = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(students_csv, schools_csv, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
df_school= pd.DataFrame(schools_csv)
print(df_school)
total_schools = df_school.groupby(['school_name'])['school_name'].count()
total_schools = df_school.groupby(['school_name'])['school_name'].count().sum()
total_schools
```

```python
# Calculate the total number of students
df_students= pd.DataFrame(students_csv)
df_students
df_students.groupby(['student_name'])['student_name'].count()
total_students = df_students.groupby(['student_name'])['student_name'].count().sum()
total_students
```

```python
# Calculate the total budget
total_budget =  df_school['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math = df_students['math_score'].mean()
average_math
```

```python
# Calculate the average (mean) reading score
average_reading = df_students['reading_score'].mean()
average_reading
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = df_students[(df_students['math_score'] >= 70)].count()['student_name']
passing_math_percentage = passing_math_count / float(total_students) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passeed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = df_students[(df_students['reading_score'] >= 70)].count()['student_name']
passing_reading_percentage = passing_reading_count / float(total_students) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_read_count = df_students[(df_students['math_score'] >= 70) & (df_students['reading_score'] >= 70)
].count()["student_name"]
overall_pass_rate = passing_math_read_count /  float(total_students) * 100
overall_pass_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary_df = pd.DataFrame({"Total Schools":[total_schools], "Total Students":[total_students], 
                                 "Total Budget":[total_budget], "Average Math":[average_math], "Average Read":[average_reading],
                                 "% Math":[passing_math_percentage], "% Read":[passing_reading_percentage],
                                 "Overall Pass":[overall_pass_rate]})
district_summary_df

# Formatting
district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df["Average Math"]=district_summary_df["Average Math"].map("{:,.2f}".format)
district_summary_df["Average Read"]=district_summary_df["Average Read"].map("{:,.2f}".format)
district_summary_df["% Math"]=district_summary_df["% Math"].map("{:.2f}%".format)
district_summary_df["% Read"]=district_summary_df["% Read"].map("{:.2f}%".format)
district_summary_df["Overall Pass"]=district_summary_df["Overall Pass"].map("{:,.2f}".format)
district_summary_df

# Display the DataFrame
district_summary_df
```

## School Summary

```python
# Use the code provided to select the school name
school_name = len(school_data_complete["school_name"].unique())
school_name
```

```python
# Use the code provided to select the school type
school_types = schools_csv.set_index(["school_name"])["type"]
school_types
```

```python
# Calculate the total student count
student_per_school_count = school_data_complete["school_name"].value_counts()
student_per_school_count
```

```python
# Calculate the total school budget and per capita spending
per_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
per_school_budget
```

```python
# Calculate the per capita spending

per_school_capita = per_school_budget / student_per_school_count
per_school_capita
```

```python
# Calculate the average math scores

average_math_school = school_data_complete.groupby(["school_name"]).mean()["math_score"]
average_math_school
```

```python
# Calculate the average reading scores

average_read_school = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
average_read_school
```

```python
# Calculate the number of schools with math scores of 70 or higher
school_pass_math = school_data_complete[(school_data_complete['math_score'] >= 70)]
school_pass_math                            
school_math_group = school_pass_math.groupby(["school_name"]).count()["student_name"]
school_math_group.head()
school_pass_math_percent = school_math_group / student_per_school_count * 100
school_pass_math_percent 
```

```python
# Calculate the number of schools with reading scores of 70 or higher
school_pass_read = school_data_complete[(school_data_complete['reading_score'] >= 70)]
school_pass_read                            
school_read_group = school_pass_read.groupby(["school_name"]).count()["student_name"]
school_read_group.head()
school_pass_read_percent = school_read_group / student_per_school_count * 100
school_pass_read_percent
```

```python
# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher
passing_math_and_reading = school_data_complete[(school_data_complete["reading_score"] >= 70) & 
                          (school_data_complete["math_score"] >= 70)]
passing_math_and_reading    
                                                
school_overall_group = passing_math_and_reading.groupby(["school_name"]).count()["student_name"]
school_overall_group.head()

school_overall_pass_rate = school_overall_group / student_per_school_count * 100
school_overall_pass_rate
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_pass_math.groupby(["school_name"]).count()["student_name"] / student_per_school_count * 100
per_school_passing_reading = school_pass_read.groupby(["school_name"]).count()["student_name"] / student_per_school_count * 100
overall_passing_rate = passing_math_and_reading.groupby(["school_name"]).count()["student_name"] / student_per_school_count * 100
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.

per_school_summary = pd.DataFrame({"School Type":school_types, "Total Students in school":student_per_school_count,
                                 "Total School Budget":per_school_budget, "Each Student Budget":per_school_capita,
                                 "Average Math Score":average_math_school, 
                                 "Average Reading Score":average_read_school, 
                                 "% Pass Math in School":school_pass_math_percent,
                                 "% Pass Reading in School":school_pass_read_percent, 
                                 "% Overall Pass Rate":school_overall_pass_rate})
per_school_summary
```

```python
per_school_summary["Total School Budget"]=per_school_summary["Total School Budget"].map("${:,}".format)
per_school_summary["Each Student Budget"]=per_school_summary["Each Student Budget"].map("${:,}".format)
per_school_summary["Average Math Score"]=per_school_summary["Average Math Score"].map("{:,.2f}".format)
per_school_summary["Average Reading Score"]=per_school_summary["Average Reading Score"].map("{:,.2f}".format)
per_school_summary["% Pass Math in School"]=per_school_summary["% Pass Math in School"].map("{:.2f}%".format)
per_school_summary["% Pass Reading in School"]=per_school_summary["% Pass Reading in School"].map("{:.2f}%".format)
per_school_summary["% Overall Pass Rate"]=per_school_summary["% Overall Pass Rate"].map("{:.2f}%".format)
per_school_summary
```

```python
# Making up the dataframe for spending_summary_df

school_spending_df["Average Math Score"]= pd.to_numeric(school_spending_df["Average Math Score"])
school_spending_df["Average Reading Score"]= pd.to_numeric(school_spending_df["Average Reading Score"])
school_spending_df["% Pass Math in School"]= school_spending_df["% Pass Math in School"].str.strip('%').astype(float)
school_spending_df["% Pass Reading in School"]= school_spending_df["% Pass Reading in School"].str.strip('%').astype(float)
school_spending_df["% Overall Pass Rate"]= school_spending_df["% Overall Pass Rate"].str.strip('%').astype(float)


spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Pass Math in School"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Pass Reading in School"]
overall_passing_rate = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Pass Rate"]

spending_summary_df=pd.DataFrame({"Average Math Score": spending_math_scores,
                                   "Average Reading Score": spending_reading_scores,
                                   "% Pass Math in School": spending_passing_math,
                                   "% Pass Reading in School": spending_passing_reading,
                                   "% Overall Pass Rate": overall_passing_rate
                                })

spending_summary_df.head()
```
